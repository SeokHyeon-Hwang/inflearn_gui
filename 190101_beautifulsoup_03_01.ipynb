```python
from bs4 import BeautifulSoup
import urllib.request as req
import lxml
```

### 네이버 금융

```python
url = 'http://finance.naver.com/sise/'
res = req.urlopen(url).read()
soup = BeautifulSoup(res, 'lxml')
```

```python
top = soup.select('#popularItemList > li')
print(top)
```

```python
for i in top:
    ti = i.find('a').string
    pi = i.find('span').string
    print('title =', ti, '/', 'price =', pi)
```

### 인프런 추천 강좌 목록

```python
url3 = 'https://www.inflearn.com/all-courses/'
res3 = req.urlopen(url3).read()
soup3 = BeautifulSoup(res3, 'lxml')
print(soup3.prettify())
```

```python
top3 = soup3.select('ul#course-list')
print(top3)
```

```python
top3_1=top3[0]
top3_2 = top3_1.find_all('li')
print(top3_2)
```

```python
print(len(top3_2))
```

```python
top3_3 = top3_2[0]
print(top3_3)
```

```python
top3_4 = top3_3.select_one('div.item-title').string
print(top3_4)
```

```python
top3_5 = top3_3.select('div.item-credits')[0]
print(top3_5)
```

```python
top3_6 = top3_5.select('strong > span ')[0]
print(top3_6.text)
```

### 인프런 다시

```python
url3 = 'https://www.inflearn.com/all-courses/'
res3 = req.urlopen(url3).read()
soup3 = BeautifulSoup(res3, 'lxml')
print(soup3.prettify())
```

```python
top3 = soup3.select('ul#course-list')
print(top3)
```

```python
top3_1=top3[0]
top3_2 = top3_1.find_all('li')
print(top3_2)
```

```python
for i in range(len(top3_2)):
    title = top3_2[i].select_one('div.item-title').string
    print(title)
```

```python
for i in range(len(top3_2)):
    price = top3_2[i].select('div.item-credits')[0]
    if price is not None:
        price2 = price.select('strong > span')
        print(price2)
    else:
        print('W0')
    
```
